```python
import numpy as np

data = np.genfromtxt('tut3.txt')
print(data)
```

```python
length = int(len(data))
comps = data[0:int(len(data)/2)]
swaps = data[int(len(data)/2):int(len(data))]
print(np.max(data))
```

```python
n = []
for i in range(10,26):
    n.append(2**i)
print(n)
```

```python
yLinear1 = [0, np.max(comps)]
yLinear2 = [0, np.max(swaps)]
xlinear = [0, np.max(n)]
```

```python
from matplotlib import pyplot as plt

# Set the plot size
plt.figure(figsize=(18,8))

# Add grid with solid lines to graph
plt.grid(linestyle='dashed')

# Customize Plot
plt.xlabel("Input Size $n$", fontsize='x-large')
plt.ylabel("Number of Operations", fontsize='x-large')
plt.title("Empirical Investigation of Quicksort", fontsize='xx-large')
plt.plot(n, comps, "b", label='Comparisons')
plt.plot(n, swaps, "r--", label='Swaps')
plt.plot(xlinear, yLinear1, "k", label="Linear 1")
plt.plot(xlinear, yLinear2, "g", label="Linear 2")

# Sets legend
legend = plt.legend(loc='upper left', shadow=(True), fontsize='xx-large', borderaxespad=3)
frame = legend.get_frame()
frame.set_facecolor('#f9f9f9')
frame.set_alpha(0.6)

# plt.ticklabel_format(useOffset=False)
# plt.yscale('linear')
```
