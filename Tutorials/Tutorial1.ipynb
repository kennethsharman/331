```python
import numpy as np

data = np.genfromtxt('data.txt')
print(data)
```

```python
length = int(len(data))
n_vals = data[0:int(len(data)/2)]
av_comp = data[int(len(data)/2):int(len(data))]

print(data)
print(n_vals)
print(av_comp)
```

```python
from matplotlib import pyplot as plt

# Set the plot size
plt.figure(figsize=(15,7))

# Add grid with solid lines to graph
plt.grid(linestyle='dashed')

# Customize Plot
plt.scale.LinearScale
plt.xlabel("$n$", fontsize='x-large')
plt.ylabel("Average Number of Comparisons", fontsize='x-large')
plt.title("Empirical investigation of binary search complexity", fontsize='xx-large')
plt.plot(n_vals, av_comp, "b-")
plt.show()
```
